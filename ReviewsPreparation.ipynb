<h2>Database Import</h2>

```python
reviews = pd.read_csv(r"googleplaystore_user_reviews.csv")
```

```python
display(reviews)
```

```python
reviews = reviews.drop(columns=['Sentiment', 'Sentiment_Polarity', 'Sentiment_Subjectivity'])
```

```python
reviews = reviews.dropna()
```

```python
display(reviews)
```

```python
apps = reviews["App"].unique().tolist()
appsReviews = []
```

```python
appsReviews = []
for app in apps:
    revs = []
    for index in reviews.index:
        if app == reviews.loc[index,'App']:
            revs.append(reviews.loc[index,'Translated_Review'])
    revs = " ".join(revs)
    revs = revs.split(" ")
    words = {}
    for word in revs:
            words[word] = 0
    for word in revs:
            words[word] +=1
    revs = []
    for i in range(5):
        try:
            revs.append(max(words, key=words.get))
            del words[max(words, key=words.get)]
        except:
            print(revs)
    appsReviews.append(";".join(revs))
```

```python
print(appsReviews)
```

```python
d = {'App': apps, 'reviews': appsReviews}
dfToWrite = pd.DataFrame(data=d)
dfToWrite.to_csv (r'processedReviews.csv', index = None, header=True)
```
